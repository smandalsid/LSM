```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("table 1.csv")
df
```

```python
plt.plot(df["increase in market size"],df["Jan-23"], "-.")
plt.plot(df["increase in market size"],df["Jan-22"], "-")
plt.legend(["Jan 2023", "Jan 2022"])
plt.title("Increase in market size")
plt.xlabel("Vehicle segment")
plt.ylabel("No. of Sales")
```

```python
df2=pd.read_csv("table 2.csv")
df2
```

```python
x=np.arange(len(df2))
plt.rcParams['figure.figsize'] = [8, 4]
plt.bar(x-0.2,df2["2022"], 0.4)
plt.bar(x+0.2,df2["2021"], 0.4)
plt.xticks(x, df2["expenses"])
plt.legend(["Year 2022", "Year 2021"])
plt.title("Expenses of M&M")
plt.xlabel("Year")
plt.ylabel("Amount in Crs.")
```
