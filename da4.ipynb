```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("percent_colorblind.csv")
df1
```

```python
df2=pd.read_csv("population.csv")
df2
```

```python
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
br1 = np.arange(len(df1["Males"]))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

plt.bar(br1, df1["Males"], width = barWidth, label ='Males')
plt.bar(br2, df1["Females"], width = barWidth, label ='Females')
plt.bar(br3, df1["Total"], width = barWidth, label ='Total')
plt.xticks(br2, df1["Nation"])
plt.legend()
plt.title("Percent of people having color blindness (in %)")
```

```python
barWidth = 0.25
# fig = plt.subplots(figsize =(12, 8))
plt.bar(np.arange(len(df2["Population"])), df2["Population"])
plt.xticks(np.arange(len(df2["Population"])), df2["Country"],rotation = 50)
plt.title("Potential market size in different countries(in Million people)")
```

```python

explode = ()
for i in range(len(df2)):
    explode=explode+(0.05,)
fig = plt.subplots(figsize =(12, 8))
# Pie Chart
plt.pie(df2["Population"], labels=df2["Country"],
        autopct='%1.1f%%', pctdistance=0.85,
        explode=explode)
 
# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
 
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
 
# Adding Title of chart
plt.title('Proportion of profit from different countries')
 
# Add Legends
# plt.legend(df2["Country"], loc="upper left", title="Fruits Color")
 
# Displaying Chart
plt.show()
```
